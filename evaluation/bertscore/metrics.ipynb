```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_edu = pd.read_csv('bert_scores_edu.csv')
df_adm = pd.read_csv('bert_scores_adm.csv')
df_leg = pd.read_csv('bert_scores_leg.csv')
```

```python
questions = list(range(1, 101))
f1_edu = df_edu['F1']  
f1_adm = df_adm['F1']  
f1_leg = df_leg['F1']  

plt.figure(figsize=(9, 6))

plt.plot(questions, f1_edu, label="Education and Culture")
plt.plot(questions, f1_adm, label="Public Administration and Procedures")
plt.plot(questions, f1_leg, label="Legislation and Regulation")

plt.xlabel("Question ID")
plt.ylabel("F1")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
